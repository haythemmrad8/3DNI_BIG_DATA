# importation :

```python
import cv2 
```

```python
import matplotlib.pyplot as plt 
```

# Model :

```python
configuration_file='ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'frozen_inference_graph.pb'
```

# loading :

```python
model = cv2.dnn_DetectionModel(frozen_model,configuration_file) 
```

```python
classLabels = [] #empty List of python
file_name ='labels.txt'
with open(file_name,'rt') as fpt :
    classLabels = fpt.read().rstrip('\n').split('\n')
```

```python
print(classLabels)
```

```python
#Set up the configurations :

model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5,127.5,127.5) )
model.setInputSwapRB(True)
```

# read an image ! 

```python
img1 = cv2.imread('test1.jpg')
```

```python
plt.imshow(img1)
```

```python
plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB))
```

```python
def ImgDetct(img):
    font_scale =1
    font = cv2.FONT_HERSHEY_PLAIN
    ClassIndex, confidece ,bbox = model.detect(img, confThreshold=0.55)
    for ClassInd, conf,boxes in zip(ClassIndex.flatten(),confidece.flatten(),bbox):
        cv2.rectangle(img,boxes,(0,0,255),1)
        cv2.putText(img,classLabels[ClassInd-1],(boxes[0]+5,boxes[1]+20),font,fontScale=font_scale ,color=(255,0,0),thickness=1)
        print(classLabels[ClassInd-1])
    return plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
        
```

```python
test2 = cv2.imread('test2.jfif')
```

```python
ImgDetct(test2)
```

```python
test3=cv2.imread('test3.jfif')
ImgDetct(test3)
```

```python
img7=cv2.imread("img7.jpg")
ImgDetct(img7)
```

```python
from matplotlib.pyplot import figure

figure(figsize=(40,40))

plt.subplot(2, 3, 1)
img1=cv2.imread("img1.jpg")
ImgDetct(img1)

plt.subplot(2, 3, 2)
img2=cv2.imread("img2.jpg")
ImgDetct(img2)

plt.subplot(2, 3, 3)
img3=cv2.imread("img3.jpg")
ImgDetct(img3)


plt.subplot(2, 3, 4)
img4=cv2.imread("img4.jpg")
ImgDetct(img4)

plt.subplot(2, 3, 5)
img5=cv2.imread("img5.jpg")
ImgDetct(img5)

plt.subplot(2, 3, 6)
img6=cv2.imread("img6.jpg")
ImgDetct(img6)

plt.show()
```

# video Demo 

```python
cap = cv2.VideoCapture("exp1.mp4")
#checked if the video is opened correctly 
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("cannot load video")
font_scale = 1
font = cv2.FONT_HERSHEY_PLAIN
while True : 
    ret , frame = cap.read()
    ClassIndex, confidece ,bbox = model.detect(frame, confThreshold=0.55)
    if(len(ClassIndex)!=0):
        for ClassInd, conf,boxes in zip(ClassIndex.flatten(),confidece.flatten(), bbox):
            if(ClassInd<=80):         
                cv2.rectangle(frame,boxes,(0,0,255),1)
                cv2.putText(frame,classLabels[ClassInd-1],(boxes[0]+5,boxes[1]+20),font,fontScale=font_scale ,color=(255,0,0),thickness=1)
    cv2.imshow('detection complete', frame)
    
    if cv2.waitKey(2) & 0xFF == ord(' '): # close video when press space 
        break
    
cap.release()
cv2.destroyAllWindows()
```
